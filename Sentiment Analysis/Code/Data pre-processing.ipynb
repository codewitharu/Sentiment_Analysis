```python
#Removing Duplicates tweets
df1.drop_duplicates(subset ="tweet_id",
                     keep = "first", inplace = True)
```

```python
def preprocessing_tweet_text(text):
    #lower Case
    text= text.lower()
    
    #removing urls
    text= re.sub(r'http\S+|www\S+|https\S+',"",text)
    
    #removing \n and @
    text= re.sub(r'\@\w+|\n',"",text)
    #deleting #
    text= re.sub(r'\#'," ",text)
    #removing extra space
    text = re.sub(' +', ' ', text)
    
    return text
```

```python
text_list= list(df1.text)
for i in range(len(text_list)):
    text_list[i]= preprocessing_tweet_text(text_list[i])
```

```python
df1['text']= text_list
```

```python
word_list=[]
text_list= list(df1.text)
for i in range(len(text_list)):
   
    #word Tokenization
    
    word_list.append(word_tokenize(text_list[i]))

df1['word']= word_list
```

```python
#removing Stop words
for i in range(len(word_list)):
    for words in word_list[i]:
        
        if words in ENGLISH_STOP_WORDS:
            print(words)
            word_list[i].remove(words)
           
```

```python
for i in range(len(word_list)):
    for words in word_list[i]:
        if(len(words)==1 or words.isdigit()):
            print(words)
            word_list[i].remove(words)
```

```python
for i in range(len(word_list)):
    
    for word in word_list[i]:
        if not word.isalpha():
            print(word)
            word_list[i].remove(word)
```

```python
pstemmer = PorterStemmer()
lemma = WordNetLemmatizer()
for strs in word_list:
    #Stemming
    for i in range(0, len(strs)):
        strs[i]=pstemmer.stem(strs[i])
        
     #Lemmatizing
    for i in range(0,len(strs)):
        strs[i]=lemma.lemmatize(strs[i], pos='a')   
```
